```python
import h2o
h2o.init()
```

```python
data = h2o.import_file('iris_wheader.csv')
```

```python
train, test = data.split_frame([0.8])
```

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator

estimator = H2OGradientBoostingEstimator()
fit = estimator.train(['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid'],'class',train)
```

```python
estimator
```

```python
predictions = estimator.predict(test)
```

```python
estimator.model_performance(test)
```

```python
perf = estimator.model_performance(test)
```

```python
perf.confusion_matrix()
```
